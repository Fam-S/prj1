Genre Analysis

```python
val2 = Da_Sample['Genre'].value_counts().values
idx2 = Da_Sample['Genre'].value_counts().index

plt.pie(val2 ,labels =idx2 ,autopct='%1.2f%%', colors=['c','m'])

plt.legend()
plt.title("Gerne Parts Percentage")

plt.show()
```

```python
# Bar chart for genre analysis
Da_Sample['Genre'].value_counts().plot(kind='bar' ,color ='c')

plt.title('Genre Distribution in Top 100 Bestselling Books')
plt.xlabel('Genre')
plt.ylabel('Number of Books')

plt.show()
```

```python
f  = Da_Sample[Da_Sample['Genre']=='Fiction']
nf = Da_Sample[Da_Sample['Genre']=='Non Fiction']

fcount = f['Year'].value_counts()
idx3   = fcount.index
val3   = fcount.values

nfcount = nf['Year'].value_counts()
idx4    = nfcount.index
val4    = nfcount.values

fig ,ax = plt.subplots()
ax.plot(val3,idx3, marker='o',color='c')

ax.set_xlabel('counts',fontsize=15)
ax.set_ylabel('Year',fontsize=15)

ax2=ax.twinx()
ax2.plot(val4,idx4,marker="*",color='g')

ax2.set_ylabel('Year' ,fontsize=15)
fig.legend(['Fiction','Non Fiction'])

plt.show()
```

Relations with Reviews


```python
# Scatter plot for user ratings and reviews
sizes = Da_Sample['Reviews']/600

plt.scatter(x = Da_Sample['User Rating'], y = Da_Sample['Reviews'] ,s = sizes ,alpha = 0.4 ,c = 'black')

plt.title("User Ratings vs Reviews")
plt.xlabel('User Rating')
plt.ylabel('Number of Reviews')

plt.show()
```

```python
Re_Sorted = Da_Sample.sort_values(by=['Reviews'], ascending=False).head(25)

plt.figure(figsize = (8, 6))

plt.bar(Re_Sorted['Author'], Re_Sorted['Reviews'], color='purple')

plt.title('Relation Between Author & Reviews')
plt.xlabel('Author Name')
plt.ylabel('N.Reviews')
plt.xticks(rotation=45, ha='right') 

plt.show()

```
